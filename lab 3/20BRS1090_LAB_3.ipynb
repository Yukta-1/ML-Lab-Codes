```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
train=pd.read_csv('/content/wheat_train.csv')
train.head()
```

```python
train.drop('ID',axis=1)
```

```python
X_train=train.drop("Type", axis=1, inplace=False)
y_train=train['Type']
```

```python
log_reg=LogisticRegression()
```

```python
clf = log_reg.fit(X_train,y_train)
```

```python
test=pd.read_csv('/content/wheat_test.csv')
test.head()
```

```python
test.drop('ID',axis=1)
```

```python
clf.predict(test)
```

```python
clf.score(train)
```
