```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs as data
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('segmented_customers.csv')
data.head()
```

# 2D ANALYSIS

```python
X=data.drop(['CustomerID', 'Gender', 'Annual Income (k$)', 'cluster'], axis=1)
X.head()
```

```python
y=data['cluster']
y.head()
```

```python
scores = []
kn = list(range(1,11))
for k in kn:
  model = KMeans(n_clusters=k, init='k-means++', random_state=0)
  model.fit(X)
  scores.append(abs(model.score(X)))

plt.plot(kn, scores)
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++')
y_pred=kmeans.fit_predict(X)
print(y_pred)
```

```python
plt.scatter(X['Age'],X['Spending Score (1-100)'],c=y_pred, cmap='rainbow')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black', alpha=1)
plt.legend()
plt.show()
```

# 3D ANALYSIS

```python
X=data.drop(['CustomerID', 'Gender', 'cluster'], axis=1)
X.head()
```

```python
y=data['cluster']
y.head()
```

```python
scores = []
kn = list(range(1,11))
for k in kn:
  model = KMeans(n_clusters=k, init='k-means++', random_state=0)
  model.fit(X)
  scores.append(abs(model.score(X)))

plt.plot(kn, scores)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
y_pred=kmeans.fit_predict(X)
print(y_pred)
```

```python
# 3d scatterplot using matplotlib

fig = plt.figure(figsize = (15,15))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X['Age'],X['Spending Score (1-100)'], X['Annual Income (k$)'],c=y_pred, cmap='rainbow')
centers=kmeans.cluster_centers_
ax.scatter(centers[:,0],centers[:,1],c='black', alpha=1)
ax.set_xlabel('Age of a customer-->')
ax.set_ylabel('Anual Income-->')
ax.set_zlabel('Spending Score-->')
ax.legend()
plt.show()
```
